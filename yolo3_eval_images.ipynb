```python
import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3' 
import numpy as np
import sys
import argparse
from yolo import YOLO, detect_video
from PIL import Image
import glob
import matplotlib.pyplot as plt
from pprint import pprint
```

Evaluate test images

```python
# read model
kwargs = {}
kwargs['model_path'] = 'trained_model/trained_weights_final.h5'
kwargs['anchors_path'] = 'data/yolo_anchors.txt'
kwargs['classes_path'] = 'data/classes.txt'
kwargs['model_image_size'] = (320, 320)

yolo = YOLO(**kwargs)
```

```python
PATH_TO_TRAIN_IMAGES = glob.glob('data/images/train/*.png')
PATH_TO_TRAIN_IMAGES = [i.replace('\\','/') for i in PATH_TO_TRAIN_IMAGES]

PATH_TO_TEST_IMAGES = glob.glob('data/images/test/*.png')
PATH_TO_TEST_IMAGES = [i.replace('\\','/') for i in PATH_TO_TEST_IMAGES]

pprint(PATH_TO_TEST_IMAGES)
```

```python
for image_path in PATH_TO_TRAIN_IMAGES:
    image = Image.open(image_path)
    r_image = yolo.detect_image(image)
    plt.figure(figsize=(7, 7))
    plt.imshow(r_image)
    plt.show()
    # r_image.show()
```
